```python
import pandas as pd
```

```python
raw_covid=pd.read_csv('covid-19-all.csv')
```

```python
  #replacing the null values with zeroes
raw_covid['Confirmed'].fillna(0, inplace=True)
raw_covid['Recovered'].fillna(0, inplace=True)
raw_covid['Deaths'].fillna(0, inplace=True)
raw_covid['Date'].fillna(0, inplace=True)
raw_covid.fillna("Others", inplace=True)
```

```python
raw_covid.head(1)
```

```python
data_gb=raw_covid.groupby(['Country/Region','Province/State','Date']).sum()
```

```python
data_gb.head(10)
```

```python
#creating a csv file
data_gb.to_csv('covid_19_data.csv')
```

```python
country_list=pd.read_csv('country-and-continent-codes-list-csv_csv.csv')
country_list.head()
#country_name=country_list['Country_Name']
#country_code=country_list['Two_Letter_Country_Code']
asia=[x.lower() for x in list(country_list[country_list['Continent_Name']=='Asia']['Country_Name'])]
europe=[x.lower() for x in list(country_list[country_list['Continent_Name']=='Europe']['Country_Name'])]
antarctica=[x.lower() for x in list(country_list[country_list['Continent_Name']=='Antartica']['Country_Name'])]
africa=[x.lower() for x in list(country_list[country_list['Continent_Name']=='Africa']['Country_Name'])]
North_America=[x.lower() for x in list(country_list[country_list['Continent_Name']=='North America']['Country_Name'])]
South_America=[x.lower() for x in list(country_list[country_list['Continent_Name']=='South America']['Country_Name'])]
Australia=[x.lower() for x in list(country_list[country_list['Continent_Name']=='Oceania']['Country_Name'])]
```

```python
our_con_list=pd.read_csv('covid_19_data.csv')['Country/Region']

continents=[]
for country in our_con_list:
    if str(country).lower() in asia:
        continents.append('Asia')
    elif str(country).lower() in europe:
        continents.append('Europe')
    elif str(country).lower() in antarctica:
        continents.append('Antarctica')
    elif str(country).lower() in africa:
        continents.append('Africa')
    elif str(country).lower() in North_America:
        continents.append('North America')
    elif str(country).lower() in South_America:
        continents.append('South America')
    elif str(country).lower() in Australia:
        continents.append('Australia')
    else:
        continents.append('Others')
data_gb['Continents']=continents

data_gb.to_csv('covid_19_data.csv')   # adding continents to the original data
```
